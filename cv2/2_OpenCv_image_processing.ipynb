```
import cv2
import matplotlib.pyplot as plt
from PIL import Image
from pathlib import Path
from google.colab.patches import cv2_imshow
```

# Padding (adding border to images)
## *cv2.copyMakeBorder()*
#### argumentst:
    src: input image

    top: border width at top

    bottom: border width at bottom

    left: border width at left

    right: border width at right

    borderType: 
        1. cv2.BORDER_CONSTANT: Adds constant colored border

        2. cv2.BORDER_REFLECT: Add padding same as border elements 
                                like: |abcdef|fed
                                      |xyzwpq|qpw

        3. cv2.BORDER_REFLECT_101 or cv2.BORDER_DEFAULT: 
            Add padding same as above except leaving last element 
                                like: |abcdef|edc
                                      |xyzwpq|pwz

        4. cv2.BORDER_REPLICATE: add padding by replicating last element 
                                like: aaa|abcdef|fff
                                      xxx|xyzwpq|qqq

        5. cv2.BORDER_WRAP: add padding look like sliding
                                like: def|abcdef|abc
                                      wpq|xyzwpq|xyz
        




```
img_path = "/content/drive/MyDrive/dataSet/img/shape.jpg"
```

```
img = cv2.imread(img_path)
```

```
constant = cv2.copyMakeBorder(img, 20, 20, 20, 20, cv2.BORDER_CONSTANT)

reflect = cv2.copyMakeBorder(img, 20, 20, 20, 20, cv2.BORDER_REFLECT)

reflect_101 = cv2.copyMakeBorder(img, 20, 20, 20, 20, cv2.BORDER_REFLECT_101)

replicate = cv2.copyMakeBorder(img, 20, 20, 20, 20, cv2.BORDER_REPLICATE)

wrap = cv2.copyMakeBorder(img, 20, 20, 20, 20, cv2.BORDER_WRAP)
```

```
paddings = [img, constant, reflect, reflect_101, replicate, wrap]
title = ["original", "constant", "reflect", "reflect_101", "replicate", "wrap"]
```

```
dir = 'photos'
if not Path(dir).is_dir():
    print("file not exits: create path")
    Path(dir).mkdir()
```

```
for i, img in enumerate(paddings):
    cv2.imwrite(dir + f"/{i}.jpg", img)
```

```
for img in Path(dir).iterdir():
    im = cv2.imread(str(img))
    cv2_imshow(im)
```

```
for i, (p, t) in enumerate(zip(paddings, title)):
    plt.subplot(2, 3, i+1)
    plt.imshow(p)
    plt.title(t)
```
