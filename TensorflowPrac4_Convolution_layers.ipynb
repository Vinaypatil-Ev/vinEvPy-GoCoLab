# Convolution layer in tensorflow


## 1. Dense: -> tf.keras.layers.Dense()


```
model = tf.keras.models.Sequential()
```

```
dense_layer = tf.keras.layers.Dense(4, "softmax", input_shape=(3,))
```

## Arguments:

1. units: [int] -> output shape
2. activation: [str] -> name of activation function
3. use_bias: [bool] -> whether use bias
4. kernel_initializer: -> initializer for kernel/ weights matrix
5. bias_initializer: -> intializer for bias 



```
model.add(dense_layer)
```

```
dense_layer.get_weights()
```

## 2. Conv1D: -> tf.keras.layers.Conv1D()

```
convd_layer = tf.keras.layers.Conv1D(4, 1, activation="softmax", input_shape=(28, 3))
convd_model = tf.keras.models.Sequential()
convd_model.add(convd_layer)
```

## Arguments

1. filters: [int] -> dimnesionality of output tensor
2. kernel_size: Union[int,[int], (int)] -> lenght of 1D convolution patch
3. strides: Union[int,[int], (int)]=default:1 -> stride length
4. padding: [default:"valid, "same", "causal"]:str -> 
  * valid :: no padding, 
  * same :: padding evenly(left / right / bottom / up), 
  * causal :: outtput[t] not depends on input[t+1]
5. data_format:[default:"channel_last", "channel_first"]:str -> one of "channel_last", "channel_first"
6. dilation_rate: Union[int,[int], (int)]=default:1 -> dilation_rate
7. activation: -> activation function :: tf.keras.activation
8. use_bias: [bool] -> whether use bias vector
9. bias_initilizer: -> tf.keras.initializers
10. kernel_initilizer: -> tf.keras.initializers





```
convd_layer.get_weights()
```

## 3. Conv2D -> tf.keras.layers.Conv2D
#### 2D convolutional, spatial convolution over images
#### Input shape should be 
  * input_shape=(128, 128, 3); data_format="channel_last"
  * input_shape=(3, 128, 128); data_format="channel_first"

```
conv2d_layer = tf.keras.layers.Conv2D(4, 1, activation="softmax", input_shape=(28, 28, 3))
conv2d_model = tf.keras.models.Sequential()
conv2d_model.add(conv2d_layer)
```

```
conv2d_layer.get_weights()
```

## 3.  Conv3D -> tf.keras.layers.Conv3D
#### 2D convolutional, spatial convolution over images
#### Input shape should be 
  * input_shape=(128, 128, 128, 3); data_format="channel_last"
  * input_shape=(3, 128, 128, 128); data_format="channel_first"


```
conv3d_layer = tf.keras.layers.Conv3D(4, 1, activation="softmax", input_shape=(128, 128, 128, 3))
conv3d_model = tf.keras.models.Sequential()
conv3d_model.add(conv3d_layer)
```

```
conv3d_layer.get_weights()
```
